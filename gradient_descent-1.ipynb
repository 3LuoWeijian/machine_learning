函数1:
$$f(x) = (x-1)^2$$


```python
import matplotlib.pyplot as plt

# 定义目标函数和梯度
def target_function1(x):
    return (x-1)**2

# 目标函数的梯度
def gradient(x):
    return 2*(x-1)

# 初始参数
x = 5       # 起始位置
learning_rate = 0.1     # 学习率
iterations = 30     # 迭代次数

# 用于记录每个迭代步骤的值,当前初始化
x_values = [x]
y_values = [target_function1(x)]

# 执行梯度下降算法
for i in range(iterations):
    gradient_value = gradient(x)
    x = x - learning_rate * gradient_value

    # 记录每个迭代步骤的值
    x_values.append(x)
    y_values.append(target_function1(x))

# 创建图表
plt.figure(figsize=(5, 5))
plt.plot(x_values, y_values, marker='o', linestyle='-', color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient_Descent_Optimization')
plt.grid(True)      # 显示网格线

# 显示图表
plt.show()
print("argmin(f) :",x)
print("f(x) = ",target_function1(x))
```

函数2:
$$f(x)=(x-2)^2+(y-2)^2$$

```python
import matplotlib.pyplot as plt
import numpy as np
# 定义目标函数和梯度
def target_function2(x,y):
    return (x-2)**2 + (y-2)**2

# 目标函数的梯度
def gradient(x,y):
    return 2*(x-2),2*(y-2)

# 初始参数
x = 5       # x起始位置
y = 5       # y起始位置
learning_rate = 0.1     # 学习率
iterations = 30     # 迭代次数

# 用于记录每个迭代步骤的值，当前初始化
x_values = [x]
y_values = [y]

# 梯度下降算法
for i in range(iterations):
    gradient_x,gradient_y = gradient(x,y)
    x = x - learning_rate * gradient_x
    y = y - learning_rate * gradient_y

    # 记录每个迭代步骤的值
    x_values.append(x)
    y_values.append(y)

# 绘制函数的等高线图
x_range = np.linspace(-6, 6, 400)
y_range = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(x_range, y_range)
Z = target_function2(X, Y)

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=np.logspace(-2, 2, 60), cmap='viridis')
plt.scatter(x_values, y_values, c='red', marker='*', label='Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent')
plt.legend()        # 用来标识图表中梯度下降路径的标签
plt.grid(True)      # 显示网格线
plt.colorbar()
plt.show()

# 输出最终结果
print("argmin(f) ：", x, y)
print("f(x,y) = ", target_function2(x, y))
```
